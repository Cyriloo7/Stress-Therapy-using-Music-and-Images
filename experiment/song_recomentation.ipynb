```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity



def collaborative_filtering(data, song_name):

    # Check if song_name exists in the data
    if song_name not in song_user_matrix.index:
        print(f"Song '{song_name}' not found. Recommending popular songs.")
        # Recommend popular items (e.g., top N by popularity)
        popular_tracks = data.sort_values(by='popularity', ascending=False).head(10)
        recommendations = popular_tracks[['track_name', 'popularity']]
        return recommendations

    # Compute cosine similarity matrix
    cosine_sim = cosine_similarity(song_user_matrix, song_user_matrix)

    # Get song index
    song_idx = list(song_user_matrix.index).index(song_name)

    # Get pairwise similarity scores and sort
    sim_scores = sorted(enumerate(cosine_sim[song_idx]), key=lambda x: x[1], reverse=True)[1:11]

    # Prepare empty DataFrame for recommendations
    recommendations = pd.DataFrame(columns=['track_name', 'popularity','danceability', 'energy',
       'key', 'loudness', 'mode', 'speechiness', 'acousticness',
       'instrumentalness', 'liveness', 'valence', 'tempo', 'similarity_score'])

    # Iterate through similar songs and extract data
    for other_song_idx, similarity in sim_scores:
        similar_song_name = song_user_matrix.index[other_song_idx]
        # Merge with original data to get song names
        similar_song_info = data[data['track_name'] == similar_song_name][['track_name', 'popularity','danceability', 'energy',
       'key', 'loudness', 'mode', 'speechiness', 'acousticness',
       'instrumentalness', 'liveness', 'valence', 'tempo']].copy()
        similar_song_info['similarity_score'] = similarity
        # Select relevant columns and sort by popularity
        recommendations = pd.concat([recommendations, similar_song_info.sort_values(by='popularity', ascending=False).head(1)], ignore_index=True)
    return recommendations



def content_based_filtering(data, title):
    
    # Get the index of the song that matches the title
    idx = data[data['track_name'] == title].index[0]

    # Compute the cosine similarity matrix for this song only
    cosine_sim = cosine_similarity(tfidf_matrix[idx:idx+1], tfidf_matrix)

    # Get the pairwise similarity scores of all songs with that song
    sim_scores = list(enumerate(cosine_sim[0]))

    # Sort the songs based on the similarity scores
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Get the scores of the 10 most similar songs
    sim_scores = sim_scores[1:11]

    # Get the song indices and their corresponding similarity scores
    song_indices = [i[0] for i in sim_scores]
    similarity_scores = [i[1] for i in sim_scores]

    # Get the top 10 most similar songs
    recommendations = data[['track_name', 'popularity','danceability', 'energy',
       'key', 'loudness', 'mode', 'speechiness', 'acousticness',
       'instrumentalness', 'liveness', 'valence', 'tempo']].iloc[song_indices]

    # Add a new column for similarity scores
    recommendations['similarity_score'] = similarity_scores

    return recommendations


numeric_features = ['popularity', 'duration_ms', 'explicit', 'danceability', 'energy',
       'key', 'loudness', 'mode', 'speechiness', 'acousticness',
       'instrumentalness', 'liveness', 'valence', 'tempo', 'time_signature']
# Create the pivot table
song_user_matrix = df.pivot_table(index='track_name', values=numeric_features, aggfunc=sum)



df['combined_features'] = df[['artists', 'track_name','danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo', 'time_signature', 'track_genre']].apply(
    lambda x: ' '.join(x.astype(str)), axis=1)

tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(df['combined_features'])
```

## this approach uses a rank-based method


```python
def hybrid_filtering_rank_based(data, song_name, alpha=0.4):
    # Get recommendations from collaborative filtering
    collaborative_recs = collaborative_filtering(data.copy(), song_name)
    # Rank the recommendations
    collaborative_recs['collab_rank'] = collaborative_recs['similarity_score'].rank(ascending=False)

    # Get recommendations from content-based filtering
    content_recs = content_based_filtering(data.copy(), song_name)
    # Rank the recommendations
    content_recs['content_rank'] = content_recs['similarity_score'].rank(ascending=False)

    # Combine the dataframes
    hybrid_recs = pd.concat([collaborative_recs, content_recs],axis=0)

    # Calculate a hybrid rank using the weighted average of the collaborative and content ranks
    hybrid_recs['hybrid_rank'] = alpha * hybrid_recs['collab_rank'] + (1 - alpha) * hybrid_recs['content_rank']

    # Sort by hybrid rank
    hybrid_recs = hybrid_recs.sort_values(by='hybrid_rank')

    # Select the top 10
    hybrid_recs = hybrid_recs.head(20)

    return hybrid_recs

# Example Usage (assuming 'df' is your data)
song_name = "Comedy"

recommendations = hybrid_filtering_rank_based(df.copy(), song_name)
print(f"Top recommendations for song '{song_name}':")
recommendations
```

## This approach uses a voting system where a song gets a vote if it appears in either the collaborative or content-based recommendations


```python
def hybrid_filtering_voting(data, song_name, num_recs=10):
    # Get recommendations from collaborative filtering
    collaborative_recs = collaborative_filtering(data.copy(), song_name)
    collaborative_recs['vote'] = 1

    # Get recommendations from content-based filtering
    content_recs = content_based_filtering(data.copy(), song_name)
    content_recs['vote'] = 1

    # Combine the dataframes
    hybrid_recs = pd.concat([collaborative_recs, content_recs],axis=0)

    # Group by track name and sum the votes
    hybrid_recs = hybrid_recs.groupby('track_name').sum().reset_index()

    # Sort by vote, then by similarity score
    hybrid_recs = hybrid_recs.sort_values(by=['vote', 'similarity_score'], ascending=[False, False])

    # Select the top num_recs
    hybrid_recs = hybrid_recs.head(num_recs)

    return hybrid_recs

# Example Usage (assuming 'df' is your data)
song_name = "Comedy"

recommendations = hybrid_filtering_voting(df.copy(), song_name)
print(f"Top recommendations for song '{song_name}':")
recommendations
```

## This approach uses a threshold-based method where a song is recommended if its similarity score from either the collaborative or content-based recommendations exceeds a certain threshold


```python
def hybrid_filtering_threshold(data, song_name, threshold=0.7):
    # Get recommendations from collaborative filtering
    collaborative_recs = collaborative_filtering(data.copy(), song_name)
    collaborative_recs = collaborative_recs[collaborative_recs['similarity_score'] > threshold]

    # Get recommendations from content-based filtering
    content_recs = content_based_filtering(data.copy(), song_name)
    content_recs = content_recs[content_recs['similarity_score'] > threshold]

    # Combine the dataframes
    hybrid_recs = pd.concat([collaborative_recs, content_recs],axis=0)

    # Remove duplicates and keep the one with the highest similarity score
    hybrid_recs = hybrid_recs.sort_values('similarity_score', ascending=False).drop_duplicates('track_name').sort_index()

    return hybrid_recs

# Example Usage (assuming 'df' is your data)
song_name = "Comedy"

recommendations = hybrid_filtering_threshold(df.copy(), song_name)
print(f"Top recommendations for song '{song_name}':")
recommendations

```

## This approach uses a max-based method where the final similarity score for each song is the maximum of the similarity scores from the collaborative and content-based recommendations


```python
def hybrid_filtering_max(data, song_name):
    # Get recommendations from collaborative filtering
    collaborative_recs = collaborative_filtering(data.copy(), song_name)
    #collaborative_recs = collaborative_recs[['track_name', 'similarity_score']]

    # Get recommendations from content-based filtering
    content_recs = content_based_filtering(data.copy(), song_name)
    #content_recs = content_recs[['track_name', 'similarity_score']]

    # Combine the dataframes
    hybrid_recs = pd.concat([collaborative_recs, content_recs])

    # Group by track name and take the max similarity score
    hybrid_recs = hybrid_recs.groupby('track_name').max().reset_index()

    # Sort by similarity score
    hybrid_recs = hybrid_recs.sort_values(by='similarity_score', ascending=False)

    return hybrid_recs

# Example Usage (assuming 'df' is your data)
song_name = "Comedy"

recommendations = hybrid_filtering_max(df.copy(), song_name)
print(f"Top recommendations for song '{song_name}':")
recommendations
```

## This approach uses a min-based method where the final similarity score for each song is the minimum of the similarity scores from the collaborative and content-based recommendations


```python
def hybrid_filtering_min(data, song_name):
    # Get recommendations from collaborative filtering
    collaborative_recs = collaborative_filtering(data.copy(), song_name)
    #collaborative_recs = collaborative_recs[['track_name', 'similarity_score']]

    # Get recommendations from content-based filtering
    content_recs = content_based_filtering(data.copy(), song_name)
    #content_recs = content_recs[['track_name', 'similarity_score']]

    # Combine the dataframes
    hybrid_recs = pd.concat([collaborative_recs, content_recs])

    # Group by track name and take the min similarity score
    hybrid_recs = hybrid_recs.groupby('track_name').min().reset_index()

    # Sort by similarity score
    hybrid_recs = hybrid_recs.sort_values(by='similarity_score', ascending=False)

    return hybrid_recs[1:11]

# Example Usage (assuming 'df' is your data)
song_name = "Comedy"

recommendations = hybrid_filtering_min(df.copy(), song_name)
print(f"Top recommendations for song '{song_name}':")
recommendations
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()

def preprocess_data(data):

    data['Genres'] = label_encoder.fit_transform(data['Genres'])
    return data
```

```python
import pandas as pd
from sklearn.neighbors import NearestNeighbors

def kNN_song_recommender(processed_data, input_features, num_recs=100):
    
    # Initialize k-NN model
    model = NearestNeighbors(n_neighbors=num_recs, metric='cosine')
    model.fit(processed_data)

    # Find indices of similar songs
    input_features_df = pd.DataFrame(input_features, index=[0])
    _, indices = model.kneighbors(input_features_df.values)

    # Get top similar songs
    top_songs_index = indices[0]
    top_songs = df1.iloc[top_songs_index]

    return top_songs


# Preprocess the data
processed_data = preprocess_data(df)
input_features['Genres'] = label_encoder.fit_transform([input_features['Genres']])
# Get recommendations based on input features
recommended_songs = kNN_song_recommender(processed_data, input_features)

print("Top 10 recommended songs:")
recommended_songs
```

## Cosine Similarity with Nearest Neighbors (using sklearn)

### This model leverages cosine similarity to find similar songs. It's useful for high-dimensional data like song features.


```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity

def cosine_similarity_recommender(processed_data, input_features, num_recs=100):
    input_features_df = pd.DataFrame(input_features, index=[0])
    
    # Compute cosine similarity between input and all data points
    similarities = cosine_similarity(input_features_df.values, processed_data)
    similarities = similarities.flatten()
    
    # Get indices of the most similar songs
    top_indices = similarities.argsort()[-num_recs:][::-1]
    
    # Retrieve the top similar songs
    top_songs = df1.iloc[top_indices]
    
    return top_songs


# Preprocess the data
processed_data = preprocess_data(df)
input_features['Genres'] = label_encoder.fit_transform([input_features['Genres']])
# Get recommendations based on cosine similarity
recommended_songs_cosine = cosine_similarity_recommender(processed_data, input_features)
print("Top 10 recommended songs (Cosine Similarity):")
recommended_songs_cosine.head(20)
```

## Singular Value Decomposition (SVD) is often used in recommendation systems to reduce dimensionality and find latent factors.


```python
import pandas as pd
from sklearn.decomposition import TruncatedSVD
from sklearn.metrics.pairwise import cosine_similarity

def svd_recommender(processed_data, input_features, num_recs=100):
    # Ensure the input features match the structure of the processed data
    input_features_df = pd.DataFrame(input_features, index=[0])
    
    # Reorder input features to match the order of processed_data columns
    input_features_df = input_features_df[processed_data.columns]

    # SVD reduces dimensionality
    svd = TruncatedSVD(n_components=15, random_state=42)
    svd_data = svd.fit_transform(processed_data)
    
    # Transform the input features
    svd_input = svd.transform(input_features_df)
    
    # Compute cosine similarity in the reduced space
    similarities = cosine_similarity(svd_input, svd_data)
    similarities = similarities.flatten()
    
    # Get the indices of the most similar songs
    top_indices = similarities.argsort()[-num_recs:][::-1]
    
    # Retrieve the top similar songs
    top_songs = df1.iloc[top_indices]
    
    return top_songs

# Get recommendations based on SVD
recommended_songs_svd = svd_recommender(processed_data, input_features)
print("Top 10 recommended songs (SVD):")
recommended_songs_svd.head(20)
```
